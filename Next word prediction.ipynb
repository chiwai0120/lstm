```python
import torch
import data
import model
import torch.nn as nn
from torch.autograd import Variable
```

```python
def model_load(fn):
    with open(fn, 'rb') as f:
        model, criterion, optimizer = torch.load(f, map_location=torch.device('cpu'))
    return model
        
model = model_load('WT2.pt')
corpus = data.Corpus('data/wikitext-2')
```

```python
def getTokens(sentence):
    tokens = sentence.split()
    return tokens
```

```python
def predict_next_word(sentence):
    tokens = getTokens(sentence)
    idxs = [corpus.dictionary.word2idx[x] if x in corpus.dictionary.word2idx else corpus.dictionary.word2idx['<unk>'] for x in tokens]
    hidden = model.init_hidden(1)
    input = torch.LongTensor([idxs])
    input = input.view(1, -1).t().contiguous()
    output, hidden = model(input, hidden)
    logits = model.decoder(output)
    word_weights = output.squeeze().data.exp().cpu()
    word_idx = torch.multinomial(word_weights, 1)[:3]
    word = [corpus.dictionary.idx2word[x] for x in word_idx]
    return word
```

```python
sentence = input('Input some words:\n')
print(predict_next_word(sentence))
```
